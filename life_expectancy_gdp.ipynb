```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_data.csv")
df.head()
```

```python
df.info()
df.describe()
df['Country'].unique()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Year", y="Life expectancy at birth (years)", hue="Country")
plt.title("Life Expectancy Over Time by Country")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Year", y="GDP", hue="Country")
plt.title("GDP Over Time by Country")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = df, x="GDP", y="Life expectancy at birth (years)", hue="Country")
plt.title("GDP vs Life Expectancy by Country")
plt.xlabel('GDP')
plt.ylabel('Life expectancy')
plt.show()
```

```python
df.groupby("Country")["Life expectancy at birth (years)"].mean()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df["Life expectancy at birth (years)"], bins=20, kde=True)
plt.title("Distribution of Life Expectancy")
plt.show()
```

### English

During this project, I explored the relationship between GDP and life expectancy using data from six countries over time. I learned how to load, explore, and visualize real-world data using pandas, matplotlib, and seaborn.

The results mostly met my expectations. In general, countries with higher GDP tend to have higher life expectancy. Developed countries such as the United States and Germany show consistently high life expectancy values, while countries with lower GDP, such as Zimbabwe, have significantly lower life expectancy.

However, the analysis also shows that GDP growth does not always result in large increases in life expectancy. For example, Chile has relatively low GDP compared to major economies, yet its life expectancy is high and stable. This suggests that factors beyond GDP, such as healthcare quality and social conditions, play an important role.

Overall, the key takeaway is that while GDP and life expectancy are positively related, the relationship is not strictly linear and depends on additional socioeconomic factors.
